```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
# The path to our CSV file
life_expectancy_data_df = pd.read_csv ("../Resources/Life Expectancy Data.csv")
life_expectancy_two_data_df = pd.read_csv ("../Resources/Life Expectancy Data.csv")
# Check the first few rows of file
life_expectancy_data_df.head()
```

GDP and Health Expenditure 

```python

gdp_df=merged_df.loc[:,["percentage expenditure","GDP"]]
gdp_df=pd.DataFrame.dropna(gdp_df,how="any")
gdp_df.plot(kind="scatter",x="GDP",y="percentage expenditure")
x=gdp_df["GDP"]
y=gdp_df["percentage expenditure"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
print(f"The r-value is: {rvalue}")
line_eq=f"y = {round(slope,2)}x + {round(intercept,2)}"
regress_values = x * slope + intercept
plt.plot(x,regress_values,"r-")
plt.annotate(line_eq, ((x.min()+100,y.max()-5000)), fontsize=15, color="red")
plt.title("GDP and Health Expenditure(%)")
```

```python
# adding the immunization data into the dataframe 
chart_two=life_expectancy_data_df.groupby('Country')[['percentage expenditure','Hepatitis B','Diphtheria ','Polio']].mean()
immun_health = chart_two.rename(columns = {"percentage expenditure - Expenditure on health as a percentage of Gross Domestic Product per capita(%)": "Percent Expenditure"})
immun_health.head()
```

```python
# sorting the bottom 5 
immun_health_df = pd.DataFrame(immun_health)
immun_health_df = immun_health.sort_values(by="percentage expenditure")
immun_health_df.head()
```

```python
# bottom 5 
bottom_immun = immun_health_df.head()
bottom_immun.head()
```

```python
# top countries immunization rates 
immun_health_df.tail()
```

```python
# sorting highest to lowest within the top countries 
top_expn_df = immun_health_df.tail()
top_expn_df = top_expn_df.sort_values(by='percentage expenditure', ascending=False)
top_expn_df.tail()
```

```python

# Creating the bar graph to represent the countries with the highest percentage expenditure 

# The countries index
countries = top_expn_df.index.tolist()  

barWidth = 0.25
# Adjusting the range to match number of countries
r1 = np.arange(len(countries))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Plotting the graph with the information on csv
plt.bar(r1, top_expn_df['Hepatitis B'], color='#AC2D47', width=barWidth, edgecolor='grey', label='Hepatitis B')
plt.bar(r2, top_expn_df['Diphtheria '].values, color='#394C88', width=barWidth, edgecolor='grey', label='Diphtheria')  # Ensure column name matches, including spaces
plt.bar(r3, top_expn_df['Polio'], color='#9E9FCD', width=barWidth, edgecolor='grey', label='Polio')

# Fixing the x-ticks and creating the labels 
# Set the position of the bar in the middle of the group bars
plt.ylabel('Immunization (%)', fontweight='bold')
plt.xlabel('Country', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(countries))], countries, rotation=90)  # Rotate for better readability if needed

# Legend
plt.legend()

# Title
plt.title('Health Expenditure - Top 5 Countries')

# Show
plt.show()
```
